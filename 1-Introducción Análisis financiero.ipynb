## Creación de variables

```python
#Creamo nuevas variables para NFLX
nflx = data
nflx["NFLX2"] = nflx["NFLX"].shift(-1)  # la segunda fila de NFLX es ahora la primera de NFLX2
nflx["DifPrecioNFLX"] = nflx["NFLX"] - nflx["NFLX2"] #Creamos una columna con diferencia de precio ( precio de mañana - precio de hoy )
nflx["RetornoDiario"] = nflx["DifPrecioNFLX"] / nflx["NFLX"] #Creamos retorno diario

#nflx["Direccion"]=[ 1 if data.loc[ei, "DifPrecioNFLX"] > 0 else -1 for ei in data.index] 
nflx["Direccion"]=[ 1 if nflx.loc[ei, "DifPrecioNFLX"] > 0 else -1 for ei in nflx.index] #Creamos dirección positiva o negativa respecto al retorno diario

nflx["Average3"] = nflx["NFLX"] + nflx["NFLX"].shift(1) + nflx["NFLX"].shift(2) / 3  #Creamos promedio movil
nflx["MA40"] = nflx["NFLX"].rolling(40).mean() # promedio movil de 40 dias
nflx["MA200"] = nflx["NFLX"].rolling(200).mean() # promedio movil de 200 dias
print(nflx)
```

```python
# Run the following code to show the price difference on 2015-01-05
print('Price difference on {} is {}. direction is {}'.format('2015-01-05', nflx['DifPrecioNFLX'].loc['2015-01-05'], nflx['Direccion'].loc['2015-01-05']))
```
