# Association between two random variables

En la vida real, muchas veces nos interesa la asociación entre dos o más variables aleatorias. Un ejemplo de esto sería analizar si existe un patrón entre el cambio en el precio de las acciones y el número de días completos en los últimos cinco días. Este tipo de análisis es valioso, ya que los traders pueden usar estos patrones para tomar decisiones de compra o venta.

Para medir la asociación entre dos variables, usamos la covarianza, que nos dice si las variables tienden a aumentar o disminuir juntas. Sin embargo, la covarianza puede ser difícil de interpretar por sí sola, ya que depende de las variaciones de las variables. Para obtener una medida que indique únicamente la fuerza de la asociación, utilizamos el coeficiente de correlación. Este valor oscila entre -1 y 1: un valor cercano a 1 indica una fuerte correlación positiva, mientras que un valor cercano a -1 indica una fuerte correlación negativa.

Es importante destacar que tanto la covarianza como la correlación solo son útiles para identificar asociaciones lineales entre variables. Si sospechamos que la relación entre las variables es no lineal, podemos utilizar gráficos de dispersión para detectar patrones no lineales.

Cuando necesitamos analizar la asociación entre una variable y varias otras, usamos la regresión lineal múltiple. Esta técnica permite entender cómo una variable puede predecir o estar relacionada con varias otras de manera simultánea.

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# Import the housing information for analysis 

data = pd.read_csv('C:/Users/Usuario/Desktop/proyectos/Analisis_Financiero_1/Datasets/portfolio_data.csv', index_col=0, parse_dates=True)
data.head()
```

```python
# Use covariance to calculate the association
data.cov()
```

```python
# Use correlation to calculate the association is more appropriate in this case
data.corr()
```

```python
# scatter matrix plot
from pandas.plotting import scatter_matrix                      # histograma para mostrar la distribución de los valores
data = scatter_matrix(data, figsize=(5, 5)) # diagonal='kde'
```

## Observe the association between AMZN and NFLX:

```python
# What is the association between AMZN and NFLX you observed?
data.plot(kind='scatter', x='AMZN', y='NFLX', figsize=(5,5))
plt.show
```

El gráfico de dispersión entre AMZN (Amazon) y NFLX (Netflix) muestra una fuerte relación lineal positiva. Esto significa que, en general, cuando el precio de AMZN sube, el precio de NFLX también tiende a subir.

Algunas observaciones:

✅ La relación parece bastante lineal, lo que sugiere que podrían estar correlacionadas.

✅ Hay cierta dispersión en los valores más altos, lo que indica que la relación no es perfectamente lineal.

✅ Una correlación positiva fuerte podría significar que ambas acciones responden a factores similares del mercado.

**Fuerza de esta relación, puedes calcular el coeficiente de correlación de Pearson**

 Si el valor está cerca de 1, la relación es muy fuerte. Si está más cerca de 0, la relación es débil.

```python
correlation = data['AMZN'].corr(data['NFLX'])
print(f'Correlación entre AMZN y NFLX: {correlation}')
```
