**Resumen sobre Regresi√≥n Lineal M√∫ltiple**

#### 1. Concepto de Regresi√≥n Lineal M√∫ltiple
A diferencia de la **Regresi√≥n Lineal Simple**, donde se usa un √∫nico predictor, en la **Regresi√≥n Lineal M√∫ltiple** se utilizan m√∫ltiples predictores para estimar la variable respuesta.

La f√≥rmula matem√°tica general es:
$$ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n + \varepsilon $$
Donde:
-  *Y* es la variable dependiente (respuesta).
-  *X_1, X_2, ..., X_n* son las variables predictoras.
-  $\beta_0$ es la intersecci√≥n (valor cuando todas las \( X \) son 0).
-  $\beta_1, \beta_2, ..., \beta_n$ son los coeficientes de regresi√≥n.
-  $\varepsilon$ es el t√©rmino de error.

#### 2. Evaluaci√≥n del Modelo: Multicolinealidad y Ajuste
**Multicolinealidad**
Se presenta cuando dos o m√°s predictores est√°n altamente correlacionados, afectando la estabilidad de los coeficientes de regresi√≥n. En mercados financieros, esto es com√∫n porque los √≠ndices burs√°tiles est√°n interrelacionados.

- **Impacto:** No reduce el poder predictivo, pero hace que los coeficientes cambien dr√°sticamente ante peque√±as variaciones en los datos.
  
-  **C√≥mo Detectar y Solucionar Multicolinealidad**
  
üìå 1. Factor de Inflaci√≥n de Varianza (VIF - Variance Inflation Factor) Una m√©trica para medir cu√°n colineadas est√°n las variables. Si 10
VIF>10, hay un problema de multicolinealidad.

üìå 2. Matriz de Correlaci√≥n
Si dos o m√°s predictores tienen una correlaci√≥n mayor a 0.8 o 0.9, es una se√±al de alerta.

üìå 3. Eliminar una de las variables redundantes
Si dos predictores aportan informaci√≥n muy similar, eliminar uno puede mejorar la estabilidad del modelo.

üìå 4. Transformaciones o PCA (An√°lisis de Componentes Principales)
M√©todos como PCA permiten combinar predictores correlacionados en nuevas variables menos redundantes.


#### 3. M√©tricas de Evaluaci√≥n

1. **Error Cuadr√°tico Medio (RMSE)**
   $$ RMSE = \sqrt{\frac{\sum (Y_{real} - Y_{predicho})^2}{n-k-1}} $$
   Donde:
   - $Y_{real}$es el valor real.
   - $Y_{predicho}$ es el valor estimado por el modelo.
   - $n$ es el n√∫mero de observaciones.
   - $k$  es el n√∫mero de predictores.
   - $RMSE $ mide el error promedio en la predicci√≥n.

 
2. **Coeficiente de Determinaci√≥n Ajustado $R^2$**  
   $$ R^2_{ajustado} = 1 - \left( \frac{(1-R^2)(n-1)}{n-k-1} \right) $$
   - Penaliza la inclusi√≥n de predictores innecesarios.
   - Indica qu√© porcentaje de la variabilidad de la respuesta es explicado por el modelo.

#### 4. Interpretaci√≥n de RMSE

El **RMSE (Root Mean Squared Error)** mide el error promedio que comete el modelo al predecir la variable dependiente $Y$.

$\underline{Criterios:}$
- **RMSE bajo**: Es bueno, porque significa que las predicciones del modelo son cercanas a los valores reales.
- **RMSE alto**: Es malo, porque significa que las predicciones del modelo est√°n lejos de los valores reales.

¬øQu√© es un "RMSE bajo"?
No hay un valor universal para RMSE que sea "bueno" o "malo", ya que depende de la escala de la variable dependiente $Y$. Por ejemplo:
- Si $Y$ est√° en una escala de 0 a 100, un RMSE de 5 podr√≠a considerarse bueno.
- Si $Y$ est√° en una escala de 0 a 1, un RMSE de 0.5 ser√≠a muy alto.

$\underline{Regla  general:}$
- Compara el RMSE con la desviaci√≥n est√°ndar de $Y$. Si el RMSE es mucho menor que la desviaci√≥n est√°ndar, el modelo es √∫til.


#### 5. Interpretaci√≥n de $R^2$ ajustado

El **$R^2$ ajustado** mide cu√°nto de la variabilidad de $Y$ es explicada por el modelo, ajustado por el n√∫mero de predictores.

Criterios:
- **$R^2$ ajustado cercano a 1**: Es bueno, porque el modelo explica la mayor parte de la variabilidad de $Y$.
- **$R^2$ ajustado cercano a 0**: Es malo, porque el modelo no explica la variabilidad de $Y$.
- **$R^2$ ajustado negativo**: Es muy malo, porque el modelo es peor que simplemente usar la media de $Y$ como predicci√≥n.

Valores t√≠picos:
- **0.7 - 1.0**: Muy bueno.
- **0.5 - 0.7**: Aceptable.
- **0.0 - 0.5**: Malo.
- **< 0.0**: Muy malo.



**Comparaci√≥n entre Train y Test**

Al comparar las m√©tricas en los conjuntos de **Train** y **Test**, puedes evaluar si el modelo est√° sobreajustado (overfitting) o si generaliza bien.

**Criterios:**
- **$R^2$ ajustado**:
  - **Train > Test**: Es normal, pero si la diferencia es muy grande, puede indicar sobreajuste.
  - **Train ‚âà Test**: Es ideal, porque significa que el modelo generaliza bien.
  - **Train < Test**: Es raro y puede indicar un error en el modelo o en los datos.



- **RMSE**:
  - **Train < Test**: Es normal, pero si la diferencia es muy grande, puede indicar sobreajuste.
  - **Train ‚âà Test**: Es ideal, porque significa que el modelo generaliza bien.
  - **Train > Test**: Es raro y puede indicar un error en el modelo o en los datos.
